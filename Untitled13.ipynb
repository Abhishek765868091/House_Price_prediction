single inheritance

```python
class parent:
    pin="parent1234"
    def parent_method(self):
        print("this is a method of parent class")
class child(parent):
    pin="child5678"
    def child_method(self):
        print("this is a method of child class")
    def parent_method(self):
        print("this is method of child class")
parent_object=parent()
child_object=child()
print(parent_object.pin)
print(child_object.pin)
```

multilevel inheritance

```python
class grandparent:
    grand_parent_variable=50
    def greet(self):
        print("hello from grandparent")
class parent(grandparent):
    parent_variable=40
    def talk(self):
        print("talking formarent")
class child(parent):
    child_variable=20
    def play(self):
        print("palying from child")
child_object=child()
print(child_object.grand_parent_variable)
print(child_object.parent_variable)
```

hybrid inheritance

```python
class a:
    def method_a(self):
        print("method a")
class b(a):
    def method_b(self):
        print("method b")
class c(a):
    def method_c(self):
        print("method c")
class d(b,c):
    def method_d(self):
        print("method d")
print(b.__mro__)
print(Exception.__mro__)
print(ConnectionResetError.__mro__)
```

```python
def is_anagram(string1:str,string2:str)->bool:
    if not isinstance (string1,str) or not isinstance(string2,str):
        raise TypeError("input must be string value")
    string1=string1.replace(" ","").lower()
    string2=string2.replace(" ","").lower()
    
    if len(string1)!=len(string2):
        return False
    count={}
    for char in string1:
        if char in count:
            count[char]+=1
        else:
            count[char]=1
    #return count
    for char in string2:
        if char not in count:
            return False
            #break
        count[char]-=1
        if count[char]<0:
            return False
    else:
           return True
            
    
```

```python
if __name__=="__main__":
    string1="asu"
    string2="asu"
    try:
        result=is_anagram(string1,string2)
        print(result)
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
```
