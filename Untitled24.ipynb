```python
def rep_miss(grid:list):
    nums=[]
    for rows in grid:
        for num in rows:
            nums.append(num)
    i=0
    n=len(nums)
    while i<n:
        correct=nums[i]-1
        if nums[i]!=nums[correct]:
            nums[i],nums[correct]=nums[correct],nums[i]
        else:
            i+=1
    for i in range(n):
        if nums[i]!=i+1:
            return [nums[i],i+1]
```

```python
rep_miss([[1,3],[2,2]])
```

```python
def findMissingAndRepeatedValues(grid):
        nums = []

        # Flatten the grid
        for row in grid:
            for num in row:
                nums.append(num)

        i = 0
        n = len(nums)

        while i < n:
            correct = nums[i] - 1
            if nums[i] != nums[correct]:
                nums[i], nums[correct] = nums[correct], nums[i]
            else:
                i += 1

        for i in range(n):
            if nums[i] != i + 1:
                return [nums[i], i + 1]
```

```python
 findMissingAndRepeatedValues([[1,3],[2,2]])
```

```python
def cal_pow(x:float,n:int):
    if n==0:
        return 1
    if n<0:
        x=1/x
        n=-n
    res=1
    while n>0:
        if n%2==1:
            res*=x
        x*=x
        n//=2
    return res
        
        
    
```

```python
def algorithm(list1:list):
    current_sum=0
    max_sum=list1[0]
    for num in list1:
        current_sum+=num
        if current_sum>max_sum:
            max_sum=current_sum
        if current_sum<0:
            current_sum=0
    return max_sum
```

```python
def sort_arr(List1:list,List2:list,m:int,n:int):
    i=j=0
    merge=[]
    while i<m and j<n:
       if List1[i]<=List2[j]:
           merge.append(List1[i])
           i+=1
       else:
           merge.append(List2[j])
           j+=1
    while i<m:
         merge.append(List1[i])
         i+=1
    while j<n:
        merge.append(List2[j])
        j+=1
    for k in range(m+n):
       List1[k]=merge[k]
    return merge
        
        
```

```python
def search_matrix(nums:list,target:int):
    for num in nums:
        if num[-1]>=target:
            return target in num
    return False
```

```python
def maxSub(nums:list):
    left=0
    right=len(nums)-1
    max_sum=0
    while left<right:
        max_sum=max(max_sum,(min(nums[left],nums[right])*(right-left)))
        if nums[left]<nums[right]:
            left+=1
        else:
            right-=1
    return max_sum
```

```python
def color_sort(Nums:list):
    for i in range(len(Nums),0,-1):
        for j in range(0,i-1):
            if Nums[j]>Nums[j+1]:
                Nums[j],Nums[j+1]=Nums[j+1],Nums[j]
    return Nums
```

```python
def spiralorder(matrix):
    res=[]
    while matrix:
        res+=(matrix.pop(0))
        if matrix and matrix[0]:
            for row in matrix:
                res.append(row.pop())
        if matrix:
            res+=(matrix.pop())
        if matrix and matrix[0]:
            for row in matrix[::-1]:
                res.append(row.pop(0))
    return res
```

```python
def armstrong(nums:int):
    digit=str(nums)
    power=len(digit)
    total=sum(int(d)**power for d in digit)
    if nums== total:
        return True
    else:
        return False
        
```

```python
def duplicate(Num):
    s=set()
    for nums in Num:
        if nums in s:
            return nums
        s.add(nums)
    return None
```

```python
def search(matrix):
    res=[]
    while matrix:
        res+=(matrix.pop(0))
        if matrix and matrix[0]:
            for row in matrix:
                res.append(row.pop())
        if matrix:
            res+=(matrix.pop()[::-1])
        if matrix and matrix[0]:
            for row in matrix:
                res.append(row.pop(0))
    return res
```

```python
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        #Solution1 (Brute Force) [Time - O(n^2)]
        total_subarray = 0
        for i in range(len(nums)):
            sum=0
            for j in range(i, len(nums)):
                sum+=nums[j]
                if sum==k:
                    total_subarray+=1
        return total_subarray
```
