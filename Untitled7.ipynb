reverse string with oop concept                

```python
class reverse:
    def __init__(self,string):
        self.string=string
        return
    def char(self):
        if not isinstance(self.string,str):
            raise TypeError("input must be string")
        rev=""
        for char in self.string:
            rev=char +rev
        return rev
        
        
```

```python
if __name__=="__main__":
    s="not"
    
    try:
 
        result=reverse(s)
        print(result.char())
    except Exception as e:
        print("some error found in program",e)
    finally:
        print("___complete__")
```

```python
class largest_number:
    def __init__(self):
        pass
    def number(self,List,count):
        #List=[23,45,67,12,56]
        count=0
        for i in List:
            if i<count:
                return i
            else:
                return count
        
```
