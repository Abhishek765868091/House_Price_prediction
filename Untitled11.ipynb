```python
def counting_char(txt:str)->int:
    if not isinstance(txt,str):
        raise ValueError("input must be string")
    count={}
    for char in txt.replace(" ",""):
        if char in count:
            count[char]=count[char]+1
        else:
            count[char]=1
    return count
```

```python
counting_char("this is python")
```

```python
if __name__=="__main__":
    user_input="this is a python language"
    try:
        result=counting_char(user_input)
        print(result)
    except ValueError as e:
        print(e)
    except Exception as e:
        print(e)
    finally:
        print("__done__")
```

```python
class employee:
    def __init__(self,salary):
        self._salary=salary
    @property
    def salary(self):
        return self._salary
emp=employee(500)
print(emp.salary)
```
