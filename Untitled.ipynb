```python
from typing import List
def second_largest(nums:list[int])->int:
    if not isinstance(nums,list):
        raise TypeError("all value must be ib a list")
    if any(not isinstance(current_number,int) for current_number in nums):
        raise ValueError
    if len(nums)<2:
        raise ValueError("more than 2 integer in list")
    if nums[0]>nums[1]:
        largest=nums[0]
        second=nums[1]
    else:
        largest=nums[1]
        second=nums[0]
    for current_number in nums:
        if current_number>largest:
            second=largest
            largest=current_number
        elif current_number!=largest and current_number>second:
            second=current_number
    return second
```

```python
#main execution
if __name__=="__main__":
   try:
       Input=[34,45,78,90,34]
       result=second_largest(Input)
       print(result)
   except (TypeError,ValueError) as e:
       print(e)
   except Exception as e:
       print(e)
```

```python
def sum_of_digit(nums:int)->int:
    if not isinstance(nums,int):
        raise TypeError("input must be a integer")
        
    total=0
    while nums>0:
        
        total=total+nums%10
        nums=nums//10
    return total
```

```python
if __name__=="__main__":
    try:
        user_input=int(input("enter number:"))
        result=sum_of_digit(user_input)
        print(result)
    except Exception as e:
        print(e)
```

```python
def sorted(num1:list,num2:list,n:int,m:int):
    i=j=0
    merge=[]
    while i<n and j<m:
        if num1[i]<num2[j]:
            merge.append(num1[i])
            i+=1
        else:
            merge.append(num2[j])
            j+=1
    while i<n:
        merge.append(num1[i])
        i+=1
    while j<m:
        merge.append(num2[j])
        j+=1
    for k in range(m+n):
        num1[k]=merge[k]
    return merge
if __name__=="__main__":
    num1=[12,23,0,0,0,0]
    num2=[1,2,3]
    n=2
    m=3
    result=sorted(num1,num2,n,m)
    print(result)
```

```python
def s_b(Prices:list):
    max_price=0
    min_price=Prices[0]
    for Price in Prices[1:]:
        if Price<min_price:
            min_price=Price
        else:
            max_price=max(max_price,Price-min_price)
    return max_price
    
```

```python
def panch(s:str):
    while '()' in s or '{}' in s or '[]' in s:
        s=s.replace('()','').replace('{}','').replace('[]','')
    return False if len(s)!=0 else True
```

```python
# 1. Load & prepare data
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_excel(r"C:\Users\asing\Downloads\Sample - Superstore.xlsx")
X = df[["profit", "category"]]  # input
y = df["target"]                   # output

# 2. Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 3. Choose model & train
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)

# 4. Predict
predictions = model.predict(X_test)

# 5. Evaluate
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test, predictions))
```

```python
df = pd.read_excel(r"C:\Users\asing\Downloads\Sample - Superstore.xlsx")
print(df.shape)
print(df.head())
print(df.dtypes)
```

```python
cols_to_drop = ["Row ID", "Order ID", "Customer ID", "Customer Name",
                "Product ID", "Product Name", "Country", "Postal Code"]

# Sirf woh columns drop karo jo actually exist karti hain
cols_to_drop = [col for col in cols_to_drop if col in df.columns]

df.drop(columns=cols_to_drop, inplace=True)
print("Remaining columns:", df.columns.tolist())
```

```python
df.columns.tolist()
```

```python
# No unnecessary columns to drop - file already clean hai!

# Date columns se useful features nikalo
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

df["Order Month"] = df["Order Date"].dt.month
df["Order Year"] = df["Order Date"].dt.year
df["Ship Days"] = (df["Ship Date"] - df["Order Date"]).dt.days

# Original date columns drop karo
df.drop(columns=["Order Date", "Ship Date"], inplace=True)

print(df.isnull().sum())
print(df.shape)
```

```python
le = LabelEncoder()

# Teri actual column names ke hisaab se
cat_cols = ["Ship Mode", "Segment", "Country/Region", "City",
            "State/Province", "Region", "Category", "Sub-Category"]

for col in cat_cols:
    df[col] = le.fit_transform(df[col])

print(df.head())
```
