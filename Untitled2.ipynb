```python
def factorial(n):
    if n==0 or n==1:
        return 1
    return n*factorial(n-1)
```

```python
factorial(4)
```

```python
import json

FILE_NAME = "student_data.json"

def load_data():
    try:
        with open(FILE_NAME, "r") as file:
            return json.load(file)
    except FileNotFoundError:
        return {"students": []}
    except json.JSONDecodeError:
        print("Invalid JSON format")
        return {"students": []}

def save_data(data):
    with open(FILE_NAME, "w") as file:
        json.dump(data, file, indent=4)

def add_student(id, name, age):
    data = load_data()
    data["students"].append({
        "id": id,
        "name": name,
        "age": age
    })
    save_data(data)
    print("Student added successfully")

def view_students():
    data = load_data()
    for s in data["students"]:
        print(s)

# ----- MAIN -----
add_student(1, "Amit", 21)
add_student(2, "Riya", 22)
view_students()
```

```python
def sum_of_digit(n):
    abs(n)
    total=0
    while n>0:
       digit=n%10
       total=total+digit
       n=n//10
    return total
```
