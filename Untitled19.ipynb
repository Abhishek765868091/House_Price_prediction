```python
def find_intersection(list_1:list,list_2:list):
    if not isinstance(list_1,list) and not isinstance(list_2,list):
        raise TypeError ("list_1 & list_2 must be integer")
    empty=[]
    for num in list_1:
        if num in list_2 and num not in empty:
            empty.append(num)
    return empty
if __name__ == "__main__":
    try:
        list_1=[12,23,34,45,56,67,12]
        list_2=[12,34,90,87,76]
        result=find_intersection(list_1,list_2)
        print(result)
    except Exception as e:
        print(e)
```

```python
def remove_white_space(text:str):
    if not isinstance(text,str):
        raise ValueError ("text must be string")
    

    result=""
    pre_space=False
    for char in text.strip():
        if not char.isspace():
            result=result+char
            pre_space=False
        else:
            if pre_space==False:
                result=result+" "
                pre_space=True
    return result
if __name__ == "__main__":
    try:
        text="  \n this \t  is \r python  "
        outcome=remove_white_space(text)
        print(outcome)
    except TypeError as error:
        print(error)
    except Exception as e:
        print(e)
        
    
```

```python
def factorial(num:int):
    if not isinstance(num,int):
        raise TypeError ("input must be integer")
    if num<0:
        raise ValueError ("input must be greater then 0")
    result=1
    for i in range(1,num+1):
        result=i*result
    return result
if __name__ == "__main__":
    try:
        num=4
        outcome=factorial(num)
        print(outcome)
    except (TypeError,ValueError) as error:
        print(error)
    except Exception as e:
        print(e)
```

```python
def largest_number(List:list)->int:
    if not isinstance(List,list):
        raise TypeError("input must be list")
    if len(List)==0:
        raise ValueError("input must have some value")
    large=List[0]
    for num in List:
        if num>large:
            large=num
    return large
if __name__ == "__main__":
    try:
        List=[12,23,34,31,78,7]
        result=largest_number(List)
        print(result)
    except (ValueError,TypeError) as error:
        print(error)
    except Exception as e:
        print(e)
```

```python
def is_palindrome(text:str):
    if not isinstance(text,str):
        raise ValueError ("input must be str")
    rev_char=""
    cleaned_text=text.strip().lower()
    for char in cleaned_text:
        rev_char=char+rev_char
    return rev_char==cleaned_text
if __name__=="__main__":
     try:
         cleaned_text=input("enter input: ")
         result=is_palindrome(cleaned_text)
         print(result)
     except Exception as e:
         print(e)
```

```python
def is_Palindrome(text: str) -> bool:
    rev_char = ""
    cleaned_text = text.strip().lower()   # normalize input
    for char in cleaned_text:
        rev_char = char + rev_char        # build reversed string
    return rev_char == cleaned_text       # compare normalized strings
```

```python
is_Palindrome("  Mom")
```

```python
def count_vowels(text:str)->int:
    if not isinstance(text,str):
        raise TypeError("input must be str")
    vowels="aeiou"
    count=0
    for char in text.lower():
        if char in vowels:
            count+=1
    return count
if __name__ == "__main__":
    try:
        text=input("enter input:")
        result=count_vowels(text)
        print(result)
    except Exception as error:
        print(error)
    except Exception as e:
        print(e)
    
```

```python
def second_largest(List:list)->int:
    if not isinstance(List,list):
        raise TypeError("input must have in a list")
    if len(List)==0:
        raise ValueError("length of input must more than two")
    if List[0]>List[1]:
        largest=List[0]
        second=List[1]
    else:
        largest=List[1]
        second=List[0]
    for curr_num in List[2:]:
        if curr_num>largest:
            second=largest
            largest=curr_num
        elif curr_num!=largest and curr_num>second:
            second=curr_num
    return second
if __name__=="__main__":
    try:
        List=[12,34,67,23,89,554,90]
        result=second_largest(List)
        print(f"second largest is {result}")
    except (TypeError,ValueError) as error:
        print(error)
    except Exception as e:
        print(e)
        
    
        
```
