```python
def sorted(num1:list,num2:list,n:int,m:int):
    i=j=0
    merge=[]
    while i<n and j<m:
        if num1[i]<num2[j]:
            merge.append(num1[i])
            i+=1
        else:
            merge.append(num2[j])
            j+=1
    while i<n:
        merge.append(num1[i])
        i+=1
    while j<m:
        merge.append(num2[j])
        j+=1
    for k in range(m+n):
        num1[k]=merge[k]
    return merge
if __name__=="__main__":
    num1=[12,23,34,0,0,0]
    num2=[1,2,3]
    n=3
    m=3
    result=sorted(num1,num2,n,m)
    print(result)
```

```python
def isValid(s:str):
    d={'{':'}','(':')','[':']'}
    stack=[]
    for i in s:
        if i in d:
            stack.append(i)
        elif len(stack)==0 or d[stack.pop()]!=i:
            return False
    return  len(stack)==0
```

```python
def merge_sort(list1,list2,m,n):
    
    i=j=0
    merge=[]
    while i<m and j<n:
        if list1[i]<=list2[j]:
            merge.append(list1[i])
            i+=1
        else:
            merge.append(list2[j])
            j+=1
    while i<m:
        merge.append(list1[i])
        i+=1
    while j<n:
        merge.append(list2[j])
        j+=1
    for k in range(m+n):
        list1[k]=merge[k]
    return merge
    
if __name__=="__main__":
    list1=[1,2,3,0,0,0]
    list2=[4,5,6,0]
    m=3
    n=3
    result=merge_sort(list1,list2,m,n)
    print(result)
        
```

```python
def miss_rep(grid):
    nums=[]
    for row in grid:
        for num in row:
            nums.append(num)
    i=0
    n=len(nums)
    while i<n:
        correct=nums[i]-1
        if nums[i]!=nums[correct]:
            nums[i],nums[correct]=nums[correct],nums[i]
        else: 
            i+=1
    for i in range(n):
        if nums[i]!=i+1:
            return [nums[i],i+1]
if __name__=="__main__":
    grid=[[0,3],[2,1]]
    result=miss_rep(grid)
    print(result)
```

```python
miss_rep([[1,3],[2,2]])
```

```python
def Count(value1):
    dic={}
    for item in value1:
        if item  in dic:
            dic[item]+=1
        else:
            dic[item]=1
    majority=max(dic,key=dic.get)
    return majority
```

```python
Count([1,2,1,2])
```

```python
def single_num(value):
    res=0
    for num in value:
        res=res^num
    return res
```

```python
def sell_buy(prices):
    min_price=float('inf')
    max_price=0
    for price in prices:
        if price<min_price:
            min_price=price
        else:
            max_price=max(max_price,price-min_price)
    return max_price
```

```python
sell_buy([7,1,5,3,6,4])
```
