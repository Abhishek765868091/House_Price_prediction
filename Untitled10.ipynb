```python
def splitter(func):
    def wrapper():
        returnvalue=func()
        value=returnvalue.split(".")
        return value
    return wrapper
def upper_converter(func):
    def wrapper():
        returnvalue=func()
        value=returnvalue.upper()
        return value
    return wrapper
@splitter
@upper_converter
def greet():
    return "rohit.kumar"
print(greet())
```
