```python
def leap_year(year):
    if (year%4==0 and year%100 !=0) or (year%400==0):
       return True
    else:
       return False
```

```python
def is_power(base_value:int,power_value:int)->int:
    if not isinstance(base_value,int) or not isinstance(power_value,int):
        raise TypeError("input must be int")
    if power_value<=0:
        raise ValueError("input must be int")
    result=1
    for i in range(power_value):
        result=result*base_value
    return result
   
```

```python
def fact(n):
    if n==0 or n==1:
        return n
    return n*fact(n-1) 
```
