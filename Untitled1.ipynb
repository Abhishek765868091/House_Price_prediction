```python
from typing import List,Any
def remove_duplicates(List:List[Any])->List[Any]:
    if not isinstance(List,list):
        raise ValueError("value must be in a list")
    if len(List)==0:
        raise ValueError("empty list not accepted")
    if len(List)==1:
        raise ValueError("List has already one value.no duplicate value exist")
        return List
    empty=[]
    for i in List:
       if not i in empty:
           empty.append(i)
    return empty

    
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import WebDriverException, NoSuchElementException
import time

def google_search(query):
    try:
        # Configure Chrome options
        chrome_options = Options()
        chrome_options.add_argument("--start-maximized")  # Open browser maximized
        chrome_options.add_argument("--disable-blink-features=AutomationControlled")  # Avoid detection

        # Path to ChromeDriver (if not in PATH, specify full path)
        service = Service("chromedriver")

        # Initialize WebDriver
        driver = webdriver.Chrome(service=service, options=chrome_options)

        # Open Google
        driver.get("https://www.google.com")

        # Accept cookies if prompted (optional)
        try:
            consent_button = driver.find_element(By.XPATH, "//button[contains(text(),'Accept')]")
            consent_button.click()
        except NoSuchElementException:
            pass  # No consent popup

        # Find search box and enter query
        search_box = driver.find_element(By.NAME, "q")
        search_box.send_keys(query)
        search_box.send_keys(Keys.RETURN)

        time.sleep(2)  # Wait for results to load

        # Get search result titles
        results = driver.find_elements(By.XPATH, "//h3")
        print(f"Top search results for '{query}':")
        for idx, result in enumerate(results[:5], start=1):
            print(f"{idx}. {result.text}")

    except WebDriverException as e:
        print(f"WebDriver error: {e}")
    finally:
        # Close browser after 5 seconds
        time.sleep(5)
        driver.quit()

if __name__ == "__main__":
    google_search("Selenium Python tutorial")
```
