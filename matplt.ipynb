```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.style.available
```

Drawing our first pie chart

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(5,3))
slices=[60,40]
plt.pie(slices)
plt.title("Piechart")
plt.tight_layout()
plt.show()
```

How to add labels in pie chart

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(5,3))
slices=[60,40]
labels=["IT","HR"] # labels list
plt.pie(slices,labels=labels) # this is how we add labels in piechart
plt.title("Piechart")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(5,3))
slices=[60,40]
labels=["IT","HR"] # labels list
wedgeprops={"edgecolor":"black",'linewidth':2,'linestyle':'--'}
plt.pie(slices,labels=labels,wedgeprops=wedgeprops) # this is how you change edge color
plt.title("Piechart")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(5,3))
slices=[60,40,30,20]
labels=["IT","HR","ADMIN","FINANCE"] # labels list
colors=["blue","red","green","yellow"]
wedgeprops={"edgecolor":"black"}
plt.pie(slices,labels=labels,colors=colors,wedgeprops=wedgeprops) # this is how you change edge color
plt.title("Piechart")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(12,7))
slices=[60,40,30,20]
labels=["IT","HR","ADMIN","FINANCE"] # labels list
explode=[0.0,0.1,0.0,0.1]
colors=["blue","red","green","yellow"]
wedgeprops={"edgecolor":"black"}
textprops={"fontsize":11,"color": "black"}
shadow=True
autopct="%.2f%%"
startangle=90
plt.pie(slices,labels=labels,textprops=textprops,autopct=autopct,startangle=startangle,explode=explode,shadow=shadow,colors=colors,wedgeprops=wedgeprops) # this is how you change edge color
plt.title("Piechart")
plt.tight_layout()
plt.show()
```
