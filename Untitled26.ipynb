```python
def merge_sort(List1,List2):
    
    merge=[]
    i,j=0,0
    while i<len(List1) and j<len(List2):
        if List1[i]<=List2[j]:
            merge.append(List1[i])
            i+=1
        else:
            merge.append(List2[j])
            j+=1
    while i<len(List1):
        merge.append(List1[i])
        i+=1
    while j<len(List2):
        merge.append(List2[j])
        j+=1
    return merge
if __name__ =="__main__":
  try:
    List1=[1,2,3,4]
    List2=[5,6,7]
    result=merge_sort(List1,List2)
    print(result)
  except Exception as e:
      print(e)

            
```

```python
def is_sorted(List_1):
    for i in range(len(List_1)-1):
        if List_1[i] > List_1[i + 1]:
            return False
    return True
```

```python
is_sorted([1,2,3,2,10])
```

```python
def traverse(arr):
    pos=0
    for i in range(len(arr)):
        if arr[i]!=0:
            arr[pos]=arr[i]
            pos+=1
    while pos<len(arr):
        arr[pos]=0
        pos+=1
    return arr
```

```python
traverse([1,0,2,0,4,0,5])
```

```python
def push_zeros_to_end(arr):
    # Pointer for the position of non-zero elements
    pos = 0
    
    # Traverse the array
    for i in range(len(arr)):
        if arr[i] != 0:
            arr[pos] = arr[i]
            pos += 1
    
    # Fill remaining positions with zeros
    while pos < len(arr):
        arr[pos] = 0
        pos += 1
    
    return arr


# Example 1
N = 8
arr = [4, 5, 0, 1, 9, 0, 5, 0]
print(*push_zeros_to_end(arr))  # Output: 4 5 1 9 5 0 0 0

# Example 2
N = 6
arr = [6, 0, 1, 8, 0, 2]
print(*push_zeros_to_end(arr))  # Output: 6 1 8 2 0 0
```
