```python
class a:
    def __init__(self,name):
        self.name=name
    def __len__(self):
        return len(str(self.name))
    def __add__ (self,other_object):
        return self.name + other_object.name
    def __str__ (self):
        return self.name
    def __repr__(self):
        return f"a({self.name})"
a_object=a(2000)
print(len(a_object))
```

```python
def fibonacci(n:int)->int:
    if not isinstance(n,int):
        raise TypeError("input must be int")
    if n<0:
        raise ValueError("n greater than 0")
    a=0
    b=1
    for i in range(n):
        print(b,end=" ")
        a,b=b,a+b
```

```python
fibonacci(12)
```

```python
if __name__=="__main__":
    try:
        n=12
        result=fibonacci(n)
        print(result)
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
```
