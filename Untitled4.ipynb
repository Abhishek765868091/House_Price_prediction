```python
def sum_of_digit(nums:int)->int:
    """
       Args:
            input must be a integer.
       Return:
             sum of digit of given number.
        Raise:
             TypeError.
             ValueError.
    """
    if not isinstance(nums,int):
        raise TypeError("input must be a integer")
    if  nums=="": 
        raise VauleError("input must be atleast one or more")
    total=0
    while nums>0:
        digit=nums%10
        total=total+digit
        nums=nums//10
    return total
        
        
        
```

```python
#main execution
if __name__=="__main__":
    try:
        user_input=int(input("enter input:"))
        result=sum_of_digit(user_input)
        print(f"sum of digit : {result}")
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as error:
        print(error)
```
