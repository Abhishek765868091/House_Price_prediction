```python
def factorial(n: int)->int:
    if not isinstance(n,int):
        raise TypeError("input must be integer")
    if n < 0:
        raise ValueError("input must be greater than 0")
    if n==0 or n==1:
        return n
    return n*factorial(n-1)
        
```

```python
factorial(4)
```

```python
def second_largest(nums):
    largest = second = float('-inf')
    for n in nums:
        if n > largest:
            second = largest
            largest = n
        elif n != largest and n > second:
            second = n
    return second

print(second_largest([5, 9,9, 2,8]))
```

```python
class factorial:
    def __init__(self,n):
        self.n=n
    def calculate(self):
        fact=1
        for i in range(1,self.n+1):
            fact*=i
        return fact
```

```python
num=int(input("enter number:"))
obj=factorial(num)
obj.calculate()
```
