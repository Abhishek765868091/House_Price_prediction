```python
def odd_ballon(B):
    freq={}
    for color in B:
        freq[color]=freq.get(color,0)+1
    for color in B:
        if freq[color]%2==1:
            return color
    return "all are even"
```

```python
def ceasor(key,value):
    if key<0:
        return "invalid input"
    result=[]
    for ch in value:
        if ch.isupper():
            shifted=chr((ord(ch)-ord("A")+key)%26+ord("A"))
            result.append(shifted)
        elif ch.isdigit():
            shifted=chr((ord(ch)-ord("0")+key)%10+ord("0"))
            result.append(shifted)
        elif ch.islower():
            shifted=chr((ord(ch)-ord("a")+key)%26+ord("a"))
            result.append(shifted)
        elif ch=="-":
            result.append(ch)
        else:
            result.append(ch)
            
    return "".join(result)
            
```

```python
import math

# Bus stops and path distances
BusStops = ["TH", "GA", "IC", "HA", "TE", "LU", "NI", "CA"]
Path     = [800, 600, 750, 900, 1400, 1200, 1100, 1500]

def getFare(source, destination):
    
    # Convert to uppercase for case-insensitive comparison
    source      = source.upper()
    destination = destination.upper()
    
    # Check if source and destination are valid stops
    if source not in BusStops or destination not in BusStops:
        print("INVALID OUTPUT")
        return
    
    # Check if source and destination are the same
    if source == destination:
        print("INVALID OUTPUT")
        return
    
    # Find index of source and destination
    src_index  = BusStops.index(source)
    dest_index = BusStops.index(destination)
    
    # Calculate distance (circular route)
    total_distance = 0
    current = src_index
    
    while current != dest_index:
        current = (current + 1) % len(BusStops)  # Move to next stop (circular)
        total_distance += Path[current]           # Add path distance
    
    # Calculate fare
    fare = (total_distance / 200)
    fare = math.ceil(fare)                        # Ceiling the fare
    
    print(f"{float(fare)} INR")

# ── Take Input ──
source      = input("Enter Source Stop: ")
destination = input("Enter Destination Stop: ")

getFare(source, destination)
```

```python
class Queue:
    def __init__(self):
        self.queue=[]
    def is_empty(self):
        return len(self.queue)==0
    def enqueue (self,value):
        self.queue.append(value)
    def dequeue(self):
        if self.is_empty():
            return None
        self.queue.pop(0)
    def peek():
        if self.is_empty():
            return None
        self.queue(0)
q=Queue()
q.enqueue(2)
q.enqueue(4)
q.enqueue(6)
q.enqueue(8)
q.dequeue()
q.queue
```
