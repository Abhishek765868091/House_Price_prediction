```python
sort_list=sorted([45,23,56,37,79,42,85])
target=245
low=0
high=len(sort_list)-1
while low<=high:
    mid=(low+high)//2
    if sort_list[mid]==target:
        print(mid)
        break
    elif target<sort_list[mid]:
        high=mid-1
    else:
        low=mid+1
else:
    print(-1)
        
```
