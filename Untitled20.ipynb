```python
def even_num(num:int):
    if not isinstance(num,int):
        raise TypeError("input must be integer")
    if num<2:
        raise ValueError("input must be greater than 2")
    
    for i in range(2,num+1,2):
      print(i)
if __name__=="__main__":
    try:
        num=20
        result=even_num(num)
        print(result)
    except Exception as e:
        print(e)
```

```python
def change_seq(text:str):
    if not isinstance(text,str):
        raise TypeError("input must be str")
       
#text="python is fun"
    rev_text=[]
    word=""
    for char in text:
        if char!=" ":
            word=word+char
        else:
            rev_text.append(word)
            word=""
    rev_text.append(word)
    rev_sen=""
    for i in range(len(rev_text)-1,-1,-1):
        rev_sen=rev_sen+rev_text[i]
        if rev_sen!=" ":
            rev_sen=rev_sen+" "
    return rev_sen
if __name__=="__main__":
    try:
        text="python is fun"
        result=change_seq(text)
        print(result)
    except Exception as error:
        print(error)
    except Exception as e:
        print(e)
        
        
```

```python
def index_val(List:list):
    if not isinstance(List,list):
        raise TypeError("input must have in the list")
    if len(List)==0:
        raise ValueError("input length more than one")
#List=[12,23,34,45]
    target=45
    low=0
    high=len(List)-1
    while low<=high:
        mid=(low+high)//2
        if target==List[mid]:
            return mid
        elif target<List[mid]:
             high=mid-1
        else:
            low=mid+1
    else:
        return -1
if __name__=="__main__":  
    try:
        List=[23,45,56,67,78,89]
        result=index_val(List)
        print(result)
    except Exception as e:
        print(e)
            
        
```

```python
class MathOperations:
    def add(self, a, b, c=100):
        return a + b + c

# Create an instance of MathOperations
Class_object = MathOperations()

# Calling the add method with different argument counts
print(Class_object.add(5, 10))      # Uses default c=100 → 115
print(Class_object.add(5, 10, 20))  # Overrides c=20 → 35
```
