```python
import time
import threading
def task1():
    print(f"Task_1 has started please wait...")
    time.sleep(4) #simulate some work
    print(f"Task_1 compleated.")
def task2():
    print(f"Task_2 has started please wait...")
    time.sleep(4) #simulate some work
    print(f"Task_2 compleated.")
def task3():
    print(f"Task_3 has started please wait...")
    time.sleep(4) #simulate some work
    print(f"Task_3 compleated.")
thread1=threading.Thread(target=task1)
thread2=threading.Thread(target=task2)
thread3=threading.Thread(target=task3)

start_time=time.perf_counter()
thread1.start()
thread2.start()
thread3.start()

thread1.join()
thread2.join()
thread3.join()
end_time=time.perf_counter()
print("total time taken : ",(end_time-start_time)) #so now after joining it will show you total time taken
    
```

# multithreading & multiprocessing

```python
pow_value=3
base_value=3
result=1
for i in range(pow_value):
    result=result*base_value
print(result)
```

```python
def cal_power(pow_value:int,base_value:int)->int:
    if not isinstance(pow_value,int) or not isinstance(base_value,int):
        raise TypeError ("power value & base value must be int")
    if pow_value==0:
        return 1
    result=1
    for i in range(pow_value):
        result=result*base_value
    return result
```

```python
if __name__ == "__main__":
    try:
         pow_value=0
         base_value=3
         value=cal_power(pow_value,base_value)
         print(value)
    except Exception as e:
        print(e)
```

```python
def rec_factorial(num:int)->int:
    if num==0 or num==1:
        return 1
    return num*rec_factorial(num-1)
```

```python
rec_factorial(4)
```

```python
text_1="aa bbb cccc"
txt={}
for char in text_1.replace(" ",""):
    if char  in txt:
        txt[char]+=1
    else:
        txt[char]=1
```

```python
print(txt,end=" ")
```

```python
text_1 = "this is a python"
txt = {}
for char in text_1:
    if char not in txt:
        txt[char] = 1      # initialize first time
    else:
        txt[char] += 1     # increment if already exists
```

```python
print(txt,end=" ")
```
