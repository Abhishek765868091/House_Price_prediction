```python
def reversed(numbers):
    rev_list=[]
    for i in range(len(numbers)-1,-1,-1):
        rev_list.append(numbers[i])
    return rev_list
print(reversed([1,2,3]))    
```

```python
def reversed_list(numbers):
    rev_list = []
    for i in range(len(numbers)-1, -1, -1):
        rev_list.append(numbers[i])
    return rev_list

# Example
print(reversed_list([1,2,3,4,5,6]))   # Output: [6,5,4,3,2,1]
print(reversed_list([10,20,30]))      # Output: [30,20,10]
```
