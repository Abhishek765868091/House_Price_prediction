```python
def is_sorted(arr:list) ->bool:
    if not isinstance(arr,list):
        raise TypeError ("input must be a list")
    for i in range(len(arr)-1):
        if arr[i]>arr[i+1]:
            return False
    return True
if __name__ == "__main__":
    try:
        arr=[1,3,4,21]
        result=is_sorted(arr)
        print(result)
    except Exception as error:
        print(error)
    except Exception as e:
        print(e)
```

```python
def flatten_list(nested_list:list):
    result=[]
    temp_list=list(nested_list)
    while temp_list:
        item=temp_list.pop()
        if isinstance(item,list):
            temp_list.extend(item)
        else:
            result.append(item)
    return result
```

```python
def stock(prices):
    min_price=prices[0]
    max_profit=0
    for price in prices:
        if price<min_price:
            min_price=price
        else:
            max_profit=max(max_profit,price-min_price)
    return max_profit
```

```python
stock([7,1,5,3,6,4])
```

```python
def missing(List):
    num=[]
    for rows in List:
        for row in rows:
            num.append(row)
    i=0
    n=len(num)
    while i<n:
        correct=num[i]-1
        if num[i]!=num[correct]:
            num[i],num[correct]=num[correct],num[i]
        else:
            i+=1


    for i in range(n):
        if num[i]!=i+1:
            return [num[i],i+1]
        
```

```python
def single(num):
    res=0
    for nums in num:
        res=res^nums
    return res
```

```python
def power(x,n):
    if n==0:
        return 1
    if n<0:
        x=1/x
        n=-n
    result=1
    while n>0:
        if n%2==1:
            result*=x
        x*=x
        n//=2
    return result
        
```

```python
power(2,-1)
```

```python
def move_zero(lst):
    pos=0
    for i in range(len(lst)):
        if lst[i]!=0:
            lst[pos],lst[i]=lst[i],lst[pos]
            pos+=1
    return lst
            
```

```python
move_zero([4,0,5,0,1])
```

```python
def major(list1):
    count={}
    for num in list1:
        if num in count:
            count[num]+=1
        else:
            count[num]=1
    count=max(count,key=count.get)
    return count
```

```python
major([3,3,2,2,2])
```

```python
def flat(num):
    empty=[]
    for nums in num:
        for row in nums:
            empty.append(row)
    i=0
    n=len(empty)
    while i<n:
        correct= empty[i]-1
        if empty[i]!=empty[correct]:
            empty[i],empty[correct]=empty[correct],empty[i]
        else:
            i+=1
    for i in range(n):
        if empty[i]!=i+1:
            return [empty[i],i+1]
        
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Initialize SparkSession
spark = (
    SparkSession.builder
    .appName("SamplePySparkApp")
    .master("local[*]")   # run locally using all cores
    .getOrCreate()
)

# Create a DataFrame
department = spark.createDataFrame([
    {"id": 1, "name": "PySpark"},
    {"id": 2, "name": "ML"},
    {"id": 3, "name": "Spark SQL"}
])

# Show the DataFrame
department.show()

# Select specific columns
department.select("id").show()

# Filter rows
department.filter(col("id") > 1).show()

# Add a new column
department.withColumn("name_length", col("name").cast("string").substr(1, 10)).show()
```
