```python
from typing import List
def is_duplicate(List:list)->list:
    if not isinstance(List,list):
        raise TypeError("all value should be in a list")
    if len(List)<=0:
        raise ValueError("atleast some value")
    
    empty=[]
    for nums in List:
        if nums not in empty:
            empty.append(nums)
    return empty
```

```python
if __name__ == "__main__":
    Input=[2,2,3,2,4,4,5,6,6,6]
    try:
        result=is_duplicate(Input)
        print("value :",result)
    except ValueError as error:
        print(error)
    except TypeError as error:
        print(error)
    except Exception as e:
        print(e)
    finally:
        print("done all task")
```

```python
def factorial(n:int):
    if not isinstance(n,int):
        raise TypeError("input must be integer")
    if n<0:
        raise ValueError("input must be greater than 0")
    if n==0 or n==1:
        return 1
    return n* factorial(n-1)
```

```python
factorial(4)
```
